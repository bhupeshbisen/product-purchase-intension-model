```python
class Model:
    def __init__(self,x_train,x_test,y_train,y_test,scaler,model,):
        self.x_train = x_train
        self.x_test = x_test
        self.y_train = y_train
        self.y_test = y_test
        self.scaler=scaler
        self.model = model
    
    def model_training(self):
        # Data Scaling
        scaler=self.scaler()
        X_train_scaled=scaler.fit_transform(X_train)
        X_test_scaled=scaler.transform(X_test)
        
        # Model creation and validation
        model=self.model()
        scoring = ['accuracy','f1_macro', 'roc_auc']
        scores = cross_validate(model, X_train_scaled, y_train, cv=10, scoring=scoring, n_jobs=-1, verbose=0,return_estimator=True)
        # To find the depth of tree
        
        # To find the depth of each tree within each RandomForest model in cross-validation
        tree_depths = [[tree.tree_.max_depth for tree in estimator.estimators_] for estimator in scores['estimator']]

        
        # Print the scores
        #print("Accuracy: ",scores['test_accuracy'])
        #print("F1 Macro Scores: ", scores['test_f1_macro'])
        #print("ROC AUC Scores: ", scores['test_roc_auc'])
        print("Mean Accuracy: ",np.mean(scores['test_accuracy']))
        print("Mean F1 Macro Score: ", np.mean(scores['test_f1_macro']))
        print("Mean ROC AUC Score: ", np.mean(scores['test_roc_auc']))
        print("Depth of Tree: ",np.mean(tree_depths))
    
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
import numpy as np

# Assume `y_probs` contains the predicted probabilities for the positive class
# and `y_test` contains the true labels.

# List of thresholds you want to evaluate
thresholds = [0.9, 0.8, 0.7, 0.6, 0.5, 0.4, 0.3]

# Store ROC-AUC scores for each threshold
roc_auc_scores = {}

for threshold in thresholds:
    # Apply the threshold to get binary predictions
    y_pred = (y_probs >= threshold).astype(int)
    
    # Calculate ROC-AUC score for the current threshold
    fpr, tpr, _ = roc_curve(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)
    
    # Store the ROC-AUC score in the dictionary
    roc_auc_scores[threshold] = roc_auc

    print(f"Threshold: {threshold}, ROC-AUC Score: {roc_auc}")

# Optionally, you can print all the ROC-AUC scores together
print("\nROC-AUC Scores for Different Thresholds:")
for threshold, score in roc_auc_scores.items():
    print(f"Threshold: {threshold}, ROC-AUC Score: {score}")
```
